# Avatar Movies Reviews Analysis

## Introduction

In this jupyter notebook, we will try to analysis the reviews from one of the most famous movie in the world: Avatar. By using selenium and BeautifulSoup, we scrap all the movie revies from IMDB website and carry out the analyse.

## Source Code

### Import Library

First, we import the related libraries for web scrapping.

```python
from bs4 import BeautifulSoup
import requests
```

```python
urls = 'https://www.imdb.com/title/tt0499549/reviews?ref_=tt_urv'
page = requests.get(urls)
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup_list = soup.findAll('div', attrs = {'class': 'text show-more__control'} )
len(soup_list)
```

```python
all_reviews = []
for soup in soup_list:
    review = soup.text.lower()
    clean_review = removeStopwordsAndPunctuations(review)
    all_reviews.append(clean_review)

all_reviews

```

```python
i = 0
stem_words = []
while i < len(all_reviews):
    a = all_reviews[i]
    from nltk.stem import PorterStemmer   
    ps = PorterStemmer()
    for word in a:        
        b = ps.stem(word)
        stem_words.append(b)
        
    i += 1

stem_words
```

```python
i = 0
words = []
while i < len(all_reviews):
    a = all_reviews[i]    
    for word in a:        
        words.append(word)
        
    i += 1

words
```
