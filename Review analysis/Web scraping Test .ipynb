```python
from bs4 import BeautifulSoup
import requests
```

```python
urls = 'https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW'
page = requests.get(urls)
page.status_code
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
import re
soup_list = soup.findAll("tr")
soup_list.pop(0)
len(soup_list)
```

```python
movie_list = []
all_price = []

for soup in soup_list:
    movie_name = soup.find("a", attrs={"class":"a-link-normal"}).text
    movie_price = soup.find("td", attrs={"class":"a-text-right mojo-field-type-money"}).text
    movie_list.append(movie_name)
    all_price.append(movie_price)

all_price

        
   
    
    
    
```
